```python
import numpy as np
import numpy as np                     # import the numpy packages and use a shorter alising name
import matplotlib.pyplot as plt        # again import the matplotlib's pyplot packages
from scipy import signal 
```

```python
x = np.arange(-7,8, 1)
L = 3
L1 = 2
L2 = 1
y = L * np.sinc(1/L * x)
y1 = L1 *  np.sinc(1/L1 *x)
y2 = L2 * np.sinc(1/L2 * x)

#this signal is sinc(1/3 * x). We get 2 samples between every zero
plt.subplot(3,1,1)
plt.stem(x, y)

#this signal is sinc(1/2 * x). Every part of the signal is slowed down, hence, it takes 2 times as long for the 
# signal to pass. Instead of a 0 crossing at every sample, we are getting a 0 crossing at every other sample.
# This makes sense because we have slowed the signal down by multiplying it by 0.5. If we sped the signal up 
# (i.e by multiplying it by 2, for example), it would pass faster and we would need to sample even faster. 
# Here, we get one sample between every 0

plt.subplot(3,1,2)
plt.stem(x, y1)


#this signal is sinc(x), it is "normal time". We are getting 0 at every sample except the 0th sample because
#the sinc is crossing x = 0 for every sample we take.
# No samples between each 0. 
plt.subplot(3,1,3)
plt.stem(x, y2)
```

```python

#Here, the sinc actually has 2 zero crossings between each sampled zero crossing! 
y3 = np.fft.fft(np.sinc(2 * x))
plt.stem(y3)
```

```python
# play around with relationship of wavelength and frequency of a signal
fs = 25
seconds = 4
signal_freq = 0.25 #note that the wavelength of a signal is inversely proportionl to the signal_freq

t = np.linspace(0, seconds, fs * seconds)
y = np.sin(2 * np.pi * signal_freq * t)

plt.stem(t,y)

#phase: 
ph = np.arctan(1/signal_freq)
print(np.angle())
print(ph)
```
